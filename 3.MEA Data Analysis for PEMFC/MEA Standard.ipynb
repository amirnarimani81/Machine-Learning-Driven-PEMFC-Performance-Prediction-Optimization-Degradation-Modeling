<b>Nyquist Plot Analysis</b><br>
🔍 <u>Set 1 - Voltage Effects</u><br>
• Decreasing semicircle size from 0.3V (largest) to 0.8V (smallest)<br>
• Charge transfer resistance (R<sub>ct</sub>) reduces by ~75% across voltage range<br>
• Negative Z<sub>imag</sub> values indicate inductive behavior at high frequencies<br>

🔍 <u>Humidity Effects (25psig)</u><br>
• RH increase (30%→100%) reduces semicircle diameter by 60%<br>
• 100% RH shows lowest impedance (0.005Ω vs 0.035Ω at 30% RH)<br>
• Reveals dramatic hydration benefits for proton conduction<br>

<b>Polarization Curve Analysis</b><br>
⚡ <u>Performance at 30% RH, 5psig</u><br>
• Two distinct tests show similar trends but different current limits<br>
• Voltage collapses completely above 2500 mA/cm² in one test<br>
• Ohmic region dominates between 500-1500 mA/cm²<br>

⚡ <u>Multi-Set Comparison</u><br>
• Set-1 maintains highest voltage (0.9V at 500 mA/cm²)<br>
• Performance spread >0.2V between best/worst sets at 2000 mA/cm²<br>
• All sets show similar mass transport limitations >2000 mA/cm²<br>

<b>Key Technical Insights</b><br>
📌 <u>Optimal Conditions</u><br>
• Best combination: 0.7-0.8V operation at 25psig with 100% RH<br>
• At 5psig, current should be limited to <2000 mA/cm²<br>

📌 <u>Performance Factors</u><br>
• Humidity affects low-current performance most (kinetic control)<br>
• Pressure improves high-current operation (mass transport control)<br>
• Set variations suggest manufacturing consistency issues<br>

<b>Recommendations</b><br>
✅ Prioritize humidity control (RH≥50%) for baseline performance<br>
✅ Higher pressures (25psig) enable higher current operation<br>
✅ Investigate Set-1 configuration for performance replication<br>
✅ Avoid operation above 0.8V due to possible catalyst degradation<br>
