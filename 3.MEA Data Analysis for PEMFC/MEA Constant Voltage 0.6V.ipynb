<b>Nyquist Plots (EIS Analysis)</b><br>
🔬 <u>Set 2 - Voltage Variation (30% RH, 5psig)</u><br>
• Shows decreasing semicircle diameters from 0.3V→0.8V<br>
• Reveals charge transfer resistance (R<sub>ct</sub>) reduces with higher voltage<br>
• Lowest impedance at 0.8V suggests optimal operating voltage<br>

🔬 <u>Pressure Variation (30% RH, 0.3V)</u><br>
• Three distinct curves for 5psig, 15psig, and 25psig<br>
• Higher pressure = smaller semicircles (↓ R<sub>ct</sub>)<br>
• 25psig shows 60% lower impedance than 5psig<br>

<b>Polarization Curves (Performance Analysis)</b><br>
⚡ <u>Voltage-Current Relationships</u><br>
• All curves show three regions: activation, ohmic, and concentration losses<br>
• Voltage drops sharply above 1500 mA/cm² (mass transport limitations)<br>
• Set-1 performs best (maintains >0.7V at 2000 mA/cm²)<br>

⚡ <u>Pressure Effects (30% RH)</u><br>
• Higher pressure extends linear (ohmic) region<br>
• 25psig maintains voltage >0.5V at 3000 mA/cm²<br>
• 5psig shows early voltage collapse at 2000 mA/cm²<br>

<b>Key Observations</b><br>
📌 Voltage optimization: 0.7-0.8V shows best performance<br>
📌 Pressure benefits: 25psig reduces impedance by 50-60% vs 5psig<br>
📌 Current limits: Performance degrades >2000 mA/cm² at low pressure<br>
📌 Set variations: Set-1 outperforms others by 15-20%<br>

<b>Technical Interpretation</b><br>
• Smaller Nyquist arcs = better charge transfer kinetics<br>
• Lower Z<sub>real</sub> intercept = reduced ohmic resistance<br>
• Flatter polarization slope = better mass transport<br>
• Higher endpoint voltage = superior cell performance<br>
